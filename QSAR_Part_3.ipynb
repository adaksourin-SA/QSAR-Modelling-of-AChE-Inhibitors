**What is PubChem?**

PubChem is a massive chemical database maintained by the NIH. Among many things, PubChem defines a standard fingerprint scheme:

881 fingerprint bits. Each bit corresponds to a specific chemical pattern.

For example:
```bash
Bit 1  -> Aromatic ring present?
Bit 2  -> Alcohol group present?
Bit 3  -> Amine present?
...
Bit 881 -> Some specific structural pattern
```

PubChem defines what the bits mean.

**What is PaDEL?**

PaDEL-Descriptor is software that reads a molecule and calculates descriptors/fingerprints.

Think:
```bash
SMILES
   ↓
PaDEL
   ↓
Features
```

PaDEL can calculate:
```bash
Lipinski descriptors
Topological descriptors
Electronic descriptors
Geometrical descriptors
PubChem fingerprints
MACCS fingerprints
Substructure fingerprints
```
and many others.
