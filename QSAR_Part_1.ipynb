### Select and retrieve the bioactivity data for Human Acetylcholinesterase

```python
selected_target = targets['target_chembl_id'][1]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id= selected_target).filter(standard_type= "IC50")
```

```python
activities = list(res)
df = pd.DataFrame(activities)
df
```

```python
df.columns
```

```python
df['canonical_smiles'].isna().sum()
```

```python
df['standard_value'].isna().sum()
```

We don't want molecules that have missing values in `canonical_smiles` or `standard_value`

```python
df2 = df[df.canonical_smiles.notna()]
df2 = df2[df2.standard_value.notna()]
df2
```

```python
print(df2['canonical_smiles'].isna().sum())
print(df2['standard_value'].isna().sum())
```

```python
len(df2['canonical_smiles'].unique())
```

We only want molecules with unique chemical descriptors

```python
df3 = df2.drop_duplicates(['canonical_smiles'])
df3
```

We don't need so many columns for our purpose. We only need `molecule_chembl_id`, `canonical_smiles` & `standard_value`

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df4 = df3[selection]
df4
```

Let's categorise the molecules based on `standard_values` as active (<1000 nM), inactive (>1000 nM) and intermediate (in between)

```python
df4.standard_value.info()
```

```python
df4['standard_value'] = df4['standard_value'].astype("float64")
```

```python
import numpy as np
conditions = [
    df4['standard_value'] < 1000,
    (df4['standard_value'] >= 1000) &  (df4['standard_value'] <= 10000),
    df4['standard_value'] > 10000
]
choices = ['active', 'intermediate', 'inactive']

df4['class'] = np.select(conditions, choices)
```

```python
df4
```

```python
df4.reset_index(drop= True, inplace= True)
df4
```

A dot in a SMILES string indicates disconnected molecular fragments. We need to remove salts and multi-component molecules before descriptor calculation.

```python
df4[df4['canonical_smiles'].str.contains(r"\.")]
```

**NOTE:** In .str.contains(), the pattern is treated as a regular expression (regex) by default. And in regex "." means "match any single character", not a literal dot.

```python
df_no_smiles = df4.drop('canonical_smiles', axis= 1)
```

```python
smiles = []

for i in df4.canonical_smiles.tolist():
    val = str(i).split('.')
    val_longest = max(val, key= len)
    smiles.append(val_longest)

can_smiles = pd.Series(smiles, name= "canonical_smiles")
    
```

```python
df_cleaned = pd.concat([df_no_smiles, can_smiles], axis= 1)[['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'class']]
```
